# Chapter 4: Classifying Data with Tree Algorithms

## Classifying Data with Decision Trees

Decision Tree (Pohon Keputusan) adalah algoritma pembelajaran diawasi yang bekerja dengan cara mempartisi data secara berulang berdasarkan aturan keputusan (kondisi jika-maka) yang diekstrak dari fitur data. Proses pembagian ini berlanjut hingga kriteria penghentian terpenuhi, membentuk struktur hierarki mirip pohon terbalik yang terdiri dari node akar (root), node internal (split), dan node daun (leaf) yang merepresentasikan keputusan akhir. Algoritma ini sangat populer karena hasil keputusannya mudah diinterpretasikan oleh manusia. Scikit-learn menyediakan DecisionTreeClassifier untuk membangun model klasifikasi berbasis pohon ini.

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

X = np.array([[0, 0], [1, 1], [0, 1], [1, 0]])
y = np.array([0, 1, 1, 0])

model = DecisionTreeClassifier(max_depth=2, random_state=42)
model.fit(X, y)

print(model.predict([[0.5, 0.5]]))
```

## Tuning Decision Tree Hyperparameters

Model Decision Tree sangat rentan terhadap overfitting jika dibiarkan tumbuh terlalu dalam hingga menghafal setiap detail noise pada data latih. Untuk mengendalikan kompleksitas pohon dan meningkatkan kemampuan generalisasinya, kita perlu melakukan tuning pada hyperparameter model. Beberapa hyperparameter krusial yang dapat dibatasi antara lain max_depth (kedalaman maksimum pohon), min_samples_split (jumlah sampel minimum yang diperlukan untuk memisahkan node), dan min_samples_leaf (jumlah sampel minimum yang harus ada di node daun). Mengatur nilai-nilai ini membantu memangkas pohon secara efektif agar tidak terlalu kompleks.

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

X = np.array([[0, 0], [1, 1], [0, 1], [1, 0], [0.5, 0.5], [1.5, 1.5]])
y = np.array([0, 1, 1, 0, 0, 1])

model = DecisionTreeClassifier(max_depth=3, min_samples_split=4, min_samples_leaf=2, random_state=42)
model.fit(X, y)

print(model.get_depth())
```

## Ensemble Methods with Random Forests

Random Forest adalah algoritma ensemble learning kuat yang menggabungkan prediksi dari banyak pohon keputusan (Decision Trees) independen untuk menghasilkan model yang lebih stabil dan akurat. Algoritma ini menggunakan teknik bagging (Bootstrap Aggregating) untuk melatih setiap pohon pada subset acak dari data latih, dan juga mengacak subset fitur pada setiap tahap pemisahan node. Dengan menggabungkan banyak pohon yang bervariasi, Random Forest secara signifikan mampu mengurangi varians dan risiko overfitting yang biasanya menjadi kelemahan utama dari pohon keputusan tunggal. Scikit-learn menyediakan RandomForestClassifier untuk menerapkan metode ini pada tugas klasifikasi.

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier

X = np.array([[0, 0], [1, 1], [0, 1], [1, 0]])
y = np.array([0, 1, 1, 0])

model = RandomForestClassifier(n_estimators=10, max_depth=2, random_state=42)
model.fit(X, y)

print(model.predict([[0.8, 0.8]]))
```

## Feature Importance in Tree Models

Salah satu keunggulan utama dari algoritma berbasis pohon seperti Decision Tree dan Random Forest adalah kemampuannya untuk mengukur tingkat pentingnya suatu fitur (feature importance) secara otomatis. Nilai tingkat kepentingan ini dihitung berdasarkan seberapa banyak setiap fitur berkontribusi dalam mengurangi ketidakmurnian (impurity seperti Gini atau Entropy) di seluruh node pohon. Fitur yang berada di bagian atas pohon dan sering digunakan untuk memisahkan data biasanya akan mendapatkan nilai kepentingan yang lebih tinggi. Informasi ini sangat berguna untuk memahami pola data dan melakukan seleksi fitur.

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier

X = np.array([[1, 2, 100],
                 [2, 1, 105],
                 [5, 6, 98],
                 [6, 5, 102]])
y = np.array([0, 0, 1, 1])

model = RandomForestClassifier(n_estimators=10, random_state=42)
model.fit(X, y)

print(model.feature_importances_)
```
