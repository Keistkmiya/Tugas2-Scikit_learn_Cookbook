## Classifying Data with k-Nearest Neighbors

k-Nearest Neighbors (k-NN) adalah algoritma pembelajaran diawasi (supervised learning) berbasis instans yang mendasar. Model ini tidak mempelajari fungsi matematika melainkan menyimpan seluruh data latih di memori. Ketika ada data baru yang ingin diprediksi, k-NN akan menghitung jarak Euclidean antara data baru tersebut dengan semua sampel data latih, menemukan sejumlah k tetangga terdekat, lalu melakukan voting mayoritas untuk menentukan label kelasnya. Scikit-learn menyediakan KNeighborsClassifier untuk mengimplementasikan metode klasifikasi ini.

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

X = np.array([[1, 2], [1, 4], [2, 3], [5, 6], [4, 7], [5, 8]])
y = np.array([0, 0, 0, 1, 1, 1])

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

print(model.predict([[3, 4]]))
```

## Regression with k-Nearest Neighbors

Selain digunakan untuk masalah klasifikasi, algoritma k-Nearest Neighbors juga dapat diterapkan pada tugas regresi untuk memprediksi nilai kontinu. Konsep dasarnya tetap sama, yaitu mencari sejumlah k tetangga terdekat berdasarkan metrik jarak. Perbedaannya terletak pada cara menentukan hasil akhir: bukan melalui voting mayoritas, melainkan dengan menghitung nilai rata-rata (mean) dari target numerik para tetangga terdekat tersebut. Scikit-learn menyediakan KNeighborsRegressor untuk kebutuhan ini.

```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1.5, 3.0, 4.5, 6.0, 7.5])

model = KNeighborsRegressor(n_neighbors=2)
model.fit(X, y)

print(model.predict([[2.5]]))
```

## Radius Neighbors Classification

Radius Neighbors Classifier adalah variasi dari algoritma k-NN yang menentukan tetangga terdekat berdasarkan radius atau jarak absolut yang tetap, bukan berdasarkan jumlah k tertentu. Pendekatan ini sangat berguna ketika kepadatan data di dalam ruang fitur bervariasi secara signifikan. Jika sebuah titik data baru berada di area yang sangat padat, model akan mempertimbangkan banyak tetangga di sekitarnya. Sebaliknya, jika data berada di area yang jarang, model hanya mengandalkan sedikit tetangga yang benar-benar dekat dalam radius tersebut, sehingga lebih kokoh terhadap anomali atau outlier. Scikit-learn menyediakan RadiusNeighborsClassifier untuk mengimplementasikan teknik ini.

```python
import numpy as np
from sklearn.neighbors import RadiusNeighborsClassifier

X = np.array([[1, 2], [1, 3], [2, 2], [6, 7], [7, 6], [7, 7]])
y = np.array([0, 0, 0, 1, 1, 1])

model = RadiusNeighborsClassifier(radius=2.0)
model.fit(X, y)

print(model.predict([[1.5, 2.5]]))
```
