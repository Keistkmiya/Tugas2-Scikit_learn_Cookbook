# Chapter 1: Preprocessing Data

## Scaling a Feature to a Range

Proses scaling bertujuan untuk mengubah nilai dari fitur numerik ke dalam rentang tertentu, biasanya antara 0 dan 1. Hal ini sangat penting karena banyak algoritma machine learning yang performanya sensitif terhadap perbedaan skala antar fitur. Scikit-learn menyediakan MinMaxScaler untuk melakukan tugas ini dengan rumus (x - min) / (max - min).

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

data = np.array([[-1, 2],
                 [-0.5, 6],
                 [0, 10],
                 [1, 18]])

scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

print(data_scaled)
```

## Standardizing Features

Standardisasi atau penggantian skala varians mengubah distribusi data sehingga memiliki nilai rata-rata (mean) sama dengan 0 dan standar deviasi sama dengan 1. Proses ini sangat direkomendasikan untuk algoritma yang mengasumsikan data terdistribusi normal, seperti Support Vector Machines (SVM) atau Linear Regression. Scikit-learn menggunakan StandardScaler untuk menghitung rata-rata dan standar deviasi pada data latih agar bisa diterapkan pada data baru.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

data = np.array([[-1, 2],
                 [-0.5, 6],
                 [0, 10],
                 [1, 18]])

scaler = StandardScaler()
data_standardized = scaler.fit_transform(data)

print(data_standardized)
```

## Normalizing Features

Normalisasi menskalakan setiap sampel individu (baris data) agar memiliki norma unit. Berbeda dengan standardisasi yang bekerja per kolom (fitur), normalisasi bekerja per baris. Proses ini sering digunakan pada analisis teks atau clustering, di mana arah atau proporsi antar fitur lebih penting daripada nilai absolutnya. Scikit-learn menyediakan Normalizer dengan pilihan norma L1 atau L2.

```python
import numpy as np
from sklearn.preprocessing import Normalizer

data = np.array([[-1, 2],
                 [-0.5, 6],
                 [0, 10],
                 [1, 18]])

normalizer = Normalizer(norm='l2')
data_normalized = normalizer.fit_transform(data)

print(data_normalized)
```

## Binarizing Features

Binalisasi adalah proses mengubah data numerik kontinu menjadi nilai biner (0 atau 1) berdasarkan ambang batas (threshold) tertentu. Semua nilai yang berada di bawah atau sama dengan ambang batas akan diubah menjadi 0, sedangkan nilai di atas ambang batas akan diubah menjadi 1. Teknik ini berguna ketika nilai kualitatif keberadaan suatu fitur lebih penting daripada kuantitas spesifiknya. Scikit-learn menyediakan Binarizer untuk melakukan transformasi ini secara efisien.

```python
import numpy as np
from sklearn.preprocessing import Binarizer

data = np.array([[-1, 2],
                 [-0.5, 6],
                 [0, 10],
                 [1, 18]])

binarizer = Binarizer(threshold=0.0)
data_binarized = binarizer.fit_transform(data)

print(data_binarized)
```

## Handling Categorical Features

Banyak algoritma machine learning tidak dapat memproses data kategori berupa teks secara langsung dan membutuhkan input numerik. Salah satu teknik untuk menangani data kategori nominal adalah One-Hot Encoding, di mana setiap nilai kategori unik diubah menjadi kolom baru bernilai biner (0 atau 1). Scikit-learn menyediakan OneHotEncoder untuk mengonversi data string atau label kategori menjadi representasi matriks biner yang siap digunakan oleh model.

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder

data = np.array([['Cold'],
                 ['Warm'],
                 ['Hot'],
                 ['Cold']])

encoder = OneHotEncoder(sparse_output=False)
data_encoded = encoder.fit_transform(data)

print(data_encoded)
```

## Imputing Missing Values

Dataset di dunia nyata sering kali memiliki nilai yang hilang (missing values) akibat kesalahan input atau kegagalan sistem. Sebelum melatih model, nilai yang kosong ini harus ditangani karena sebagian besar algoritma scikit-learn tidak menerima input data yang mengandung nilai kosong. Salah satu strategi dasar adalah imputasi, yaitu mengisi nilai yang hilang menggunakan nilai statistik seperti rata-rata (mean), median, atau nilai yang paling sering muncul (most frequent) dari kolom tersebut. Scikit-learn menyediakan SimpleImputer untuk mempermudah proses ini.

```python
import numpy as np
from sklearn.impute import SimpleImputer

data = np.array([[1, 2],
                 [np.nan, 6],
                 [7, np.nan],
                 [10, 12]])

imputer = SimpleImputer(strategy='mean')
data_imputed = imputer.fit_transform(data)

print(data_imputed)
```

## Generating Polynomial Features

Ketika hubungan antara fitur dan target bersifat non-linear, pembuatan fitur polinomial dapat membantu model linear menangkap pola yang lebih kompleks. Teknik ini bekerja dengan membuat kombinasi perkalian antar fitur yang ada serta memangkatkannya hingga derajat tertentu. Scikit-learn menyediakan PolynomialFeatures yang dapat secara otomatis menghasilkan kolom baru berisi interaksi antar fitur tersebut untuk meningkatkan kapasitas representasi model linear.

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

data = np.array([[2, 3],
                 [4, 5]])

poly = PolynomialFeatures(degree=2)
data_poly = poly.fit_transform(data)

print(data_poly)
```

## Custom Transformers

Scikit-learn menyediakan berbagai metode transformasi bawaan, namun terkadang kita memerlukan transformasi kustom yang spesifik untuk kebutuhan data tertentu. Dengan menggunakan FunctionTransformer, kita dapat mengubah fungsi Python biasa menjadi sebuah transformer yang kompatibel dengan alur kerja scikit-learn. Hal ini mempermudah integrasi pembersihan data atau manipulasi matematis kustom ke dalam pipeline pemrosesan data.

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer

data = np.array([[1, 4],
                 [9, 16]])

log_transformer = FunctionTransformer(np.log1p)
data_transformed = log_transformer.transform(data)

print(data_transformed)
```
