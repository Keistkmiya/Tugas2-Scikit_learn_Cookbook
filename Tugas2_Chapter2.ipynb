# Chapter 2: Linear Models

## Fitting a Linear Regression Model

Regresi Linear adalah salah satu model statistik dan machine learning paling mendasar yang digunakan untuk memprediksi nilai kontinu (target) berdasarkan satu atau lebih fitur numerik. Model ini bekerja dengan mencari garis lurus (hyperplane) terbaik yang meminimalkan jumlah kuadrat kesalahan (residual sum of squares) antara nilai prediksi dan nilai aktual. Scikit-learn menyediakan LinearRegression untuk melatih model ini menggunakan metode Ordinary Least Squares (OLS).

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
```

## Regularizing Linear Models with Ridge Regression

Ridge Regression adalah variasi dari regresi linear yang menggunakan teknik regularisasi L2 untuk mencegah terjadinya overfitting. Ketika sebuah model linear memiliki terlalu banyak fitur atau fitur yang saling berkaitan kuat (multikolinearitas), bobot koefisien cenderung menjadi sangat besar dan sensitif terhadap noise. Ridge menambahkan penalti berupa jumlah kuadrat dari seluruh nilai koefisien ke dalam fungsi kerugian (loss function). Penalti ini memaksa nilai koefisien mengecil mendekati nol, sehingga model menjadi lebih stabil dan memiliki kemampuan generalisasi yang lebih baik pada data baru.

```python
import numpy as np
from sklearn.linear_model import Ridge

X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([2, 3, 5, 6])

model = Ridge(alpha=1.0)
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
```

## Sparsifying Linear Models with Lasso Regression

Lasso Regression (Least Absolute Shrinkage and Selection Operator) adalah teknik regularisasi model linear lainnya yang menggunakan penalti L1. Berbeda dengan Ridge yang hanya mengecilkan nilai koefisien mendekati nol, Lasso menambahkan penalti berdasarkan jumlah nilai absolut dari koefisien. Karakteristik utama dari Lasso adalah kemampuannya untuk memaksa koefisien fitur yang kurang penting menjadi tepat nol. Hal ini membuat Lasso secara otomatis berfungsi sebagai metode seleksi fitur (feature selection), menghasilkan model yang lebih sederhana (sparse model) dan lebih mudah diinterpretasikan.

```python
import numpy as np
from sklearn.linear_model import Lasso

X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([2, 3, 5, 6])

model = Lasso(alpha=0.1)
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
```

## Logistic Regression for Classification

Meskipun memiliki kata 'regression' di namanya, Logistic Regression merupakan algoritma linear yang digunakan untuk menyelesaikan masalah klasifikasi, terutama klasifikasi biner (dua kelas). Algoritma ini memetakan kombinasi linear dari fitur ke dalam fungsi sigmoid, yang menghasilkan nilai probabilitas antara 0 dan 1. Jika probabilitas melewati ambang batas tertentu (biasanya 0.5), sampel akan dimasukkan ke dalam kelas positif, dan jika kurang akan dimasukkan ke dalam kelas negatif.

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

model = LogisticRegression()
model.fit(X, y)

print(model.predict([[3.5]]))
print(model.predict_proba([[3.5]]))
```

## Stochastic Gradient Descent for Classification

Stochastic Gradient Descent (SGD) adalah pendekatan optimasi yang sangat efisien untuk melatih model linear seperti Linear Regression atau Logistic Regression, terutama ketika menangani dataset berskala sangat besar. Berbeda dengan Gradient Descent konvensional yang menghitung gradien menggunakan seluruh dataset sekaligus, SGD memperbarui parameter model secara bertahap menggunakan satu sampel data acak pada setiap iterasi. Pendekatan ini membuat proses pelatihan menjadi jauh lebih cepat dan menggunakan memori yang sangat sedikit. Scikit-learn menyediakan SGDClassifier untuk menerapkan teknik optimasi ini pada berbagai masalah klasifikasi.

```python
import numpy as np
from sklearn.linear_model import SGDClassifier

X = np.array([[1, 2], [2, 1], [3, 4], [4, 3]])
y = np.array([0, 0, 1, 1])

model = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)
model.fit(X, y)

print(model.predict([[2, 2]]))
```
